# Keyword Analysis

Enter keywords and receive a table detailing the number of reports containing those particular keywords.



```python
import pandas as pd
import numpy as np

data = pd.read_csv('Reports.csv').dropna()
```

#### Choose keywords below

```python

keywords = (["fall", "fell", "floor"], 
            ["drowsy", "tired", "bed"], 
            ["bowel", "constipat", "diarrhoea", "poo", "stool"],
            ["urine", "wee", "waterworks"], 
            ["mobility", "walking", " unsteady", "balance"], 
            ["mood", "sad", "upset", "depressed"],
            ["tired", "sleep", "fatig"],
            ["bleeding", "blood", "bleed"], 
            ["appetite", "eating"], 
            ["drink", "drank", "thirst"], 
            ["pain", "hurt", "harm"], 
            ["bruise"],
            ["confusion", "confus", "muddled"], 
            ["angry", "violent"],
            ["infection", "antibiotics", "infect", "treat"],
            ["broken", "damaged"],
            ["missing", "lost", "find"],
            ["medication", "med", "drug", "prescription"], 
            ["hospital", "discharge"],
            ["concern", "difficult", "issue", "normal"],
            ["rota", "schedule", "shift"],
            ["safe", "secure"]
                   )
```

```python
categories = len(keywords)

for j in range(0, categories):

    length = len(keywords[j])             

    results = np.array([keywords[j],np.zeros(length)])

    for i in range(0, length) :
        results[1][i] = len(data[data['message_manager'].str.contains(keywords[j][i])])

    df = pd.DataFrame(results)

    print(df)
    print("")

# OPTIONAL: uncomment to export to multiple CSVs with the name of first keyword in the category
#     name = keywords[j][0]    
#     df.to_csv('%s.csv' % name)
```
